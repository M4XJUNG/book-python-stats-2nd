### 3.4.6 표본분산

- 분산 : '데이터가 평균값(기대값)에서 얼마나 떨어져 있는가'를 나타내는 지표

- 표본분산 : 표본에서 계산한 분산

image.png

- 편차 : 데이터와 평균값의 차이
$ (x_i - \bar{x}) $

- 편차제곱합 : 분산을 나타내는 식에서 분자 부분
$ \sum_{i=1}^n (x_i - \bar{x})^2 $

image.png

image.png

### 3.4.7 표본분산(실습)

```python
s2 = np.sum((fish_length - x_bar) ** 2) / n
print(s2)
```

```python
fish_length
```

```python
fish_length - x_bar
```

```python
(fish_length - x_bar) ** 2
```

```python
print(np.sum((fish_length - x_bar) ** 2))
```

```python
print(np.var(fish_length, ddof = 0))
```

### 3.4.8 불편분산

image.png

### 3.4.9 불편분산(실습)

```python
u2 = np.sum((fish_length - x_bar) ** 2) / (n - 1)
print(u2)
```

```python
print(round(u2, 3))
```

```python
# 불편분산은 np.var 함수를 사용하고 인수에 ddof=1을 지정하면 쉽게 계산.
print(round(np.var(fish_length, ddof = 1), 3))
```

np.var 함수는 ddof = 0이면 표본분산을, ddof = 1이면 불편분산을 계산한다.

어느 분산을 계산하는지 명확히 하기 위해 ddof 인수를 반드시 지정하는 것이 좋다.



---


넘파이와 판다스에 모두 var 라는 함수가 있지만, 서로 동작이 다르다.

넘파이의 var 함수는 ddof를 지정하지 않으면 표본분산을 계산한다.

반면 판다스의 var 함수(판다스의 데이터프레임에 연결해 실행한 var 함수)에서는 ddof를 지정하지 않으면 불편분산이 계산된다.

```python
np.var(fish_length_df)
```

```python
fish_length_df.var()
```

```python
fish_length_df.var(ddof = 0)
```
