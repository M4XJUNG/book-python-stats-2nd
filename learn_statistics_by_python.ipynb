### 2.4.13 데이터프레임 병합

```python
df_1 = pd.DataFrame({
    'col1' : np.array([1, 2, 3]),
    'col2' : np.array(['A', 'B', 'C'])
})
df_2 = pd.DataFrame({
    'col1' : np.array([4, 5, 6]),
    'col2' : np.array(['D', 'E', 'F'])
})
```

```python
# 두 개의 데이터프레임을 세로 방향으로 이어 붙이기
print(pd.concat([df_1, df_2]))
```

```python
# 두 개의 데이터프레임을 가로 방향으로 이어 붙이기
print(pd.concat([df_1, df_2], axis = 1))
```

### 2.4.15 데이터프레임 행에 대한 작업

```python
# 데이터프레임에서 특정 행을 가져오는 몇 가지 방법
print(sample_df.head(n = 3))
```

```python
print(sample_df.query('index == 0'))
```

```python
# 큰 따옴표와 작은 따옴표를 사용할 때 주의!
print(sample_df.query('col3 == "A"'))
```

```python
print(sample_df.query('col3 == "A" | col3 == "D"'))
```

```python
print(sample_df.query('col3 == "A" & col3 == "D"'))
```

```python
# 행과 열에 모두 조건 지정
print(sample_df.query('col3 == "A"')[['col2', 'col3']])
```
